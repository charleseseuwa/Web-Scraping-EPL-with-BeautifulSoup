# Now Scraping Data For Multiple Season and teams with For-loop

```python
years = list(range(2022, 2020, -1)) #<-- Starts with the current season and then go backward
```

```python
years #<-- The 2 seasons we 'll be scrapping'
```

```python
all_matches = []
```

```python
standings_url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
```

```python
import time

for year in years:
    data = requests.get(standings_url)
    soup = BeautifulSoup(data.text, 'lxml')
    standings_table = soup.select('table.stats_table')[0]
    
    links = [l.get('href') for l in standings_table.find_all('a')]  #<-- find all tean links and get the href property
    links = [l for l in links if '/squads/' in l]   #<-- Filter the link for the squad
    team_urls = [f"https://fbref.com{l}" for l in links]   #<-- Turn thelink from relative to absolute  links
    
    
    previous_season = soup.select('div a[class="button2 prev"]')[0].get('href')  #Click on the previous season link
    #Alter previous_season = soup.select('a.buttons.prev')[0]['href']
    
    """
    Selecting anchor tags that has the class 'prev',
    select the first one; since it returns a list,
    then get 'href' property of that one anchor tag
    """
    
    standings_url = f"https://fbref.com{previous_season}"  #<-- convert the relative link to absoule link
    
    for team_url in team_urls:
        team_name = team_url.split('/')[-1].replace('-Stats', '').replace('-', ' ') #<-- Get the team names
        
        data = requests.get(team_url)
        matches = pd.read_html(data.text, match= 'Scores & Fixtures')[0]
        
        soup = BeautifulSoup(data.text, 'lxml')
        links = [l.get('href') for l in soup.find_all('a')]
        links = [l for l in links if l and '/all_comps/shooting' in l] #<-- Find the shooting stats link
        
        data = requests.get(f"https://fbref.com{links[0]}")  #<-- using request to get the html of ther shooting page
        shooting = pd.read_html(data.text, match="Shooting")[0]  #<-- REadinfg in shooting stat to pandas
        shooting.columns = shooting.columns.droplevel()  #<-- Drop the top index level in shooting dataframe
        
        team_data = matches.merge(shooting[['Date', 'Sh', 'SoT', 'Dist', 'FK', 'PK', 'PKatt']], on = 'Date')
        
#         try:
#             team_data = matches.merge(shooting[['Date', 'Sh', 'SoT', 'Dist', 'FK', 'PK', 'PKatt']], on = 'Date')
#         except ValueError:
#             continue
            
#             """
#             Sometimes for some teams, the shooting stats aren't available. So when we actually try to merge the two
#             dataframes together, Pandas will throws an error (ValueError) because the shooting stats dataframe is empty.
#             If Pandas throws a ValueError, just continue with the loop without doing anything; i.e ignore/skip the team 
#             where the shooting stats isn't available.
#             """

        team_data = team_data[team_data['Comp'] == 'Premier League']  #<--- We re only interested in premier league games
        team_data['Season'] = year                               # <--- Adding Season columns to the dataframe
        team_data['Team'] = team_name               #<-- Adding Team colum to the dataframe
        
        """
        Team and Season columns were added because we need to be able to distinguish which team and season the data
        is from. Unlike on the real website we scrapped, there's no need to have such columns added to the table because,
        by default being on a particular page on the website, we'll know which season and year the data represents.
        So, it's very important to always preserve the information on the website you're scrapping which isn't neccessarily
        available in the specific table we're scrapping.
        """
        
        all_matches.append(team_data)  #<-- Appending the team_data to an empty_list (all_matches)
        time.sleep(1)  #<--- sleep for secs. This slows down quick we make a request & prevent us from getting blocked
        
        
```

```python
match_df = pd.concat(all_matches)  # <-- concatenate
match_df
```

```python
match_df.columns = [c.lower() for c in match_df.columns]  #<- converting the column names to lower case
```

```python
match_df.to_csv('matches2.csv', index=False)  #<-- write the data frame to csv
```
